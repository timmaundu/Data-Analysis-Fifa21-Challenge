### Matching the Height and Weight columns which have different measurements for height we work with cm for weight we work with kg


```python
#check for unique values
data['Height'].unique()
```

```python
#check for unique values
data.Weight.unique()
```

```python
#Changing All values to unique format in height(cm)
def change_height(datas):
    new_heights = []
    for val in datas:
        if str(val).endswith('cm'):
            new_heights.append(val[:-2])
        elif '"' in val:
            value = float(val[0])*12 + float(val[2])
            value = round(value*2.54)
            new_heights.append(value)
        else:
            new_heights.append(val)
    return new_heights

data['Height'] = pd.Series(change_height(data['Height'])).astype('int')
```

```python
data['Height'].unique()
```

```python
# Changing values to unique for at in weight(kg)
def change_weight(data):
    new_weights = []
    for val in data:
        if str(val).endswith('kg'):
            new_weights.append(val[:-2])
        elif 'lbs' in val:
            value = round(float(val[:3])*0.454)
            new_weights.append(value)
        else:
            new_weights.append(val)
    return new_weights

data['Weight'] = change_weight(data.Weight)
```

### Changing wage, value and Release clause to standard format


€ M as million(1000000) and K as thousand(1000) and they need to be transformed. The function below will be used for transaformation of the 3 columns

```python
data['Wage'].sample(5)
```

```python
data['Value'].sample(5)
```

```python
data['Release_Clause'].sample(5)
```

```python
#function to transform the 3 columns.
def convert(data):
    values = []
    for val in data:
        val = str(val).replace('€', '')
        if 'M' in val:
            value = float(val.split('M')[0])*1000000
            values.append(value)
        elif 'K' in val:
            value = float(val.split('K')[0])*1000
            values.append(value)
        else:
            values.append(val)
    return values  
```

```python
#Making the changes from the function result and assigning an interger type
data['Wage'] = pd.Series(convert(data['Wage'])).astype('int')
data['Value'] = pd.Series(convert(data['Value'])).astype('int')
data['Release_Clause'] = pd.Series(convert(data['Release_Clause'])).astype('int')
```

```python
def rating(data):
    rate = []
    for val in data:
        val = val[0]
        rate.append(val)
    return rate
```

### The Contract Column with inconsistencies

```python
data['Contract'].unique()
```

```python
# function to ensure the status is well known
def contract_status(data):
    status = []
    for val in data:
        if 'Loan' in val:
            status.append('Loan')
        elif 'Free' in val:
            status.append('Free')
        else:
            status.append('Active')
    return status

data['Contract_status'] = pd.Series(contract_status(data.Contract)).astype('category')
```

### Handling the Hits and Loan Date End columns
The hits column refers to the number of times the player's profile has been viewed on the website. Some entries contains 'K'while some contains missing values(000)

```python
#The hits column is cleaned with the code below
data['Hits'] = data['Hits'].astype(str).apply(lambda x: float(x.replace('K', ''))*1000 if 'K' in x else float(x))

#Filling the Nan entries with zero(0)
data['Hits'].fillna(0, inplace = True)

#Converting the column to type int 
data['Hits'] = data['Hits'].astype('int64')

data.drop(columns = 'Loan_Date_End', inplace = True)
```
